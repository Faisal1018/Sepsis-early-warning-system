```python
import pandas as pd

df = pd.read_csv("../data/dataset.csv")

data = df.copy()

data = data.drop(
    columns=[
        "EtCO2",
        "TroponinI",
        "Bilirubin_direct",
        "Fibrinogen"
    ]
)
```

```python
data.shape
```

```python
selected_features = [
    "HR",
    "O2Sat",
    "Temp",
    "SBP",
    "MAP",
    "Resp",
    "Creatinine",
    "WBC",
    "Glucose",
    "Age",
    "Gender",
    "HospAdmTime",
    "ICULOS",
    "SepsisLabel"
]

data[selected_features].isnull().mean()*100
```

```python
keep_features = [
    "HR",
    "O2Sat",
    "Temp",
    "SBP",
    "MAP",
    "Resp",
    "Creatinine",
    "WBC",
    "Glucose",
    "Age",
    "Gender",
    "HospAdmTime",
    "ICULOS",
    "SepsisLabel",
    "Patient_ID"
]

data = data[keep_features]
```

```python
data.shape
```

```python
data["Patient_ID"].nunique()
```

```python
data.groupby("Patient_ID").size().head(10)
```

```python
from sklearn.model_selection import train_test_split

patients = data["Patient_ID"].unique()

train_patients, test_patients = train_test_split(
    patients,
    test_size=0.2,
    random_state=42
)
```

```python
len(test_patients)
```

```python
len(train_patients)
```

```python
train_df = data[data["Patient_ID"].isin(train_patients)]

test_df = data[data["Patient_ID"].isin(test_patients)]
```

```python
set(train_df["Patient_ID"]).intersection(
    set(test_df["Patient_ID"])
)
```

```python
features = [
    "HR",
    "O2Sat",
    "Temp",
    "SBP",
    "MAP",
    "Resp",
    "Creatinine",
    "WBC",
    "Glucose",
    "Age",
    "Gender",
    "HospAdmTime",
    "ICULOS"
]
```

```python
X_train = train_df[features]
y_train = train_df["SepsisLabel"]

X_test = test_df[features]
y_test = test_df["SepsisLabel"]
```

```python
#X_train.shape
#X_test.shape
y_train.shape
#y_test.shape
```

```python
X_test.shape
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
```

```python
X_train = pd.DataFrame(
    imputer.fit_transform(X_train),
    columns=X_train.columns
)
```

```python
X_test = pd.DataFrame(
    imputer.transform(X_test),
    columns=X_test.columns
)
```

```python
X_train.isnull().sum().sum()
```

```python
X_test.isnull().sum().sum()
```

```python
clean_data = pd.concat(
    [X_train, y_train.reset_index(drop=True)],
    axis=1
)

clean_data.to_csv(
    "../data/processed/sepsis_clean_train.csv",
    index=False
)
```

```python
test_data = pd.concat(
    [X_test, y_test.reset_index(drop=True)],
    axis=1
)

test_data.to_csv(
    "../data/processed/sepsis_clean_test.csv",
    index=False
)
```
